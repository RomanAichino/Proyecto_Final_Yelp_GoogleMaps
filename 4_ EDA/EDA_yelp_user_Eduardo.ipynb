```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_yelp_user_eda = pd.read_csv("/content/drive/MyDrive/Otros/yelp_user.csv")
```

```python
# Visualización de correlaciones entre las variables numéricas
correlation_matrix = df_yelp_user_eda.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()
```

```python
# Distribución de valoraciones promedio ( 1 al 5)
plt.figure(figsize=(8, 6))
sns.histplot(df_yelp_user_eda['average_stars'], kde=True)
plt.title("Distribución de Valoraciones Promedio Usuarios")
plt.xlabel("Valoraciones Promedio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Distribución de elite_count ( 1 al 5)
plt.figure(figsize=(8, 6))
sns.histplot(df_yelp_user_eda['elite_count'], kde=True)
plt.title("Distribución de conteo Anios Elite")
plt.xlabel("Conteo Elite")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Relación entre 'review_count' y 'average_stars'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='review_count', y='average_stars', data=df_yelp_user_eda[df_yelp_user_eda['review_count']<1000])
plt.title("Relación entre Review Count y Estrellas Promedio")
plt.xlabel("Review Count")
plt.ylabel("average stars")
plt.show()
```

Un usuario que escribe mas resenas tiene mas oportunidades de incrementar su valoracion(stars). Sin embargo por encima de 1000 reviews existen muy pocos usuarios.

```python
# Relación entre 'elite_count' y 'review_count'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='review_count', y='elite_count', data=df_yelp_user_eda[df_yelp_user_eda['review_count']<1000])
plt.title("Relación entre Review Count y elite cpunt")
plt.xlabel("Review Count")
plt.ylabel("elite count")
plt.show()
```

EDA Tabla YELP Review

```python
df_yelp_review_eda = pd.read_csv("/content/drive/MyDrive/Otros/yelp_review_florida.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='stars_x', data=df_yelp_review_eda)
plt.title("Distribución de Estrellas en Reseñas")
plt.xlabel("Estrellas")
plt.ylabel("Cantidad")
plt.show()
```

```python
# Relación entre 'useful', 'funny' y 'cool' con 'stars'
plt.figure(figsize=(10, 6))
sns.boxplot(x='stars_x', y='useful', data=df_yelp_review_eda)
plt.title("Relación entre 'useful' y Estrellas")
plt.xlabel("Estrellas")
plt.ylabel("Useful")
plt.show()
```

```python
# Distribución de longitud de las reseñas
df_yelp_review_eda['text_length'] = df_yelp_review_eda['text'].apply(len)
plt.figure(figsize=(8, 6))
sns.histplot(df_yelp_review_eda['text_length'], bins=30)
plt.title("Distribución de Longitud de Reseñas")
plt.xlabel("Longitud de Reseña")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Relación entre 'longitud de resenas' y 'valoracion'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='stars_x', y='text_length', data=df_yelp_review_eda)
plt.title("Relación entre longitud de resena y Estrellas Promedio")
plt.xlabel("average stars")
plt.ylabel("Longitud de resenas")
plt.show()
```

```python
# Correlación entre variables numéricas
correlation_matrix = df_yelp_review_eda.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()
```

```python
user_review_count = df_yelp_review_eda['user_id'].value_counts()
plt.figure(figsize=(10, 6))
sns.histplot(user_review_count, bins=30)
plt.title("Distribución de Cantidad de Reseñas por Usuario")
plt.xlabel("Cantidad de Reseñas")
plt.ylabel("Cantidad de Usuarios")
plt.show()
```
