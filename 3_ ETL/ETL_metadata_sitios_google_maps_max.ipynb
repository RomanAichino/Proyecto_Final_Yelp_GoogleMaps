# 3. Cargamos el csv unificadoy filtramos para Florida usando la columna address, mejoraremos este filtro mas adelante.

```python
import pandas as pd

df_metada = pd.read_csv('/content/drive/MyDrive/Otros/googlemaps_metadata_sitios.csv')
```

```python
df_metada.iloc[105]
```

```python
print(df_metada.iloc[105]['address'])
print(type(df_metada.iloc[105]['address']))
```

```python
# FILTRAMOS PARA FLORIDA

metdat_florida = df_metada[df_metada['address'].str.contains('FL', na=False, case=True)]
metdat_florida = metdat_florida.copy()
metdat_florida.head(3)
```

```python
# Le damos el formato adecuado a la columna category:

import numpy as np
import ast

metdat_florida['category'] = metdat_florida['category'].apply(lambda x: ast.literal_eval(x)
                                                                if isinstance(x, str) else np.nan)

def format_category_list(category_list):
    if isinstance(category_list, list):
        return ', '.join(category_list)
    else:
        return category_list

metdat_florida['category'] = metdat_florida['category'].apply(format_category_list)
```

```python
# Realizamos un conteo del top 24 de categorias:

category = metdat_florida['category'].str.split(', ').explode()
category_counts = category.value_counts()/len(metdat_florida)

# Mostrar el conteo de 24 categorías mas importantes
category_counts.head(24)
```

```python
top_category_names = category_counts.head(24).index.tolist()
top_category_names
```

```python
# FILTRAMOS LA TABLA DE ACUERDO AL TOP 24

categories = top_category_names

conditions = metdat_florida['category'].str.contains(categories[0], na=False, case=True)

for category in categories[1:]:
    conditions = conditions | metdat_florida['category'].str.contains(category, na=False, case=True)

metdat_florida = metdat_florida[conditions]
metdat_florida
```

```python
df_unnested = metdat_florida.explode('category')

desired_categories = ['Restaurant', 'Auto repair shop', 'Beauty salon', 'Convenience store', 'Hair salon',
                      'Doctor', 'Gas station', 'Nail salon', 'ATM', 'Church', 'Barber shop', 'Dentist',
                      'Clothing store', 'Insurance agency', 'Real estate agency', 'Cell phone store',
                      'Condominium complex', 'Park', 'Massage therapist', 'Attorney', 'Bank', 'Tire shop',
                      'Used car dealer', 'Gift shop']

df_unnested = df_unnested[df_unnested['category'].isin(desired_categories)]

category_dummies = pd.get_dummies(df_unnested['category'])
df_new = pd.concat([df_unnested, category_dummies], axis=1)

metdat_florida = df_new

metdat_florida.head(3)
```

'Payments': ['Checks', 'Debit cards', 'NFC mobile payments', 'Credit cards']

'Amenities': ['Public restroom']

'Service options': ['No-contact delivery', 'Delivery', 'In-store shopping', 'Same-day delivery']

```python
metdat_florida['MISC'] = metdat_florida['MISC'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)

desired_keys = ['Payments', 'Amenities', 'Service options']

def format_misc_dict(misc_dict):
    if isinstance(misc_dict, dict):
        filtered_dict = {k: misc_dict[k] for k in desired_keys if k in misc_dict}
        formatted_misc = ', '.join([f"{key}_{', '.join(value)}" for key, value in filtered_dict.items()])
        return formatted_misc
    else:
        return misc_dict

# Applying function
metdat_florida['MISC'] = metdat_florida['MISC'].apply(format_misc_dict)

# Create binary indicator columns for each MISC
category_dummies = metdat_florida['MISC'].str.get_dummies(sep=', ')
df_new = pd.concat([metdat_florida, category_dummies], axis=1)

metdat_florida = df_new
metdat_florida.head(3)
```

```python
# COLUMNA HOURS

print(metdat_florida.iloc[3]['hours'])
print(type(metdat_florida.iloc[3]['hours']))
```

```python
import ast
# De <class 'str'> a <class 'list'>

metdat_florida['hours'] = metdat_florida['hours'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# De <class 'list'> a <class 'dict'>

metdat_florida['hours'] = metdat_florida['hours'].apply(lambda x: dict(x) if isinstance(x, list) else x)

# De <class 'dict'> a columnas

hours_df = metdat_florida['hours'].apply(pd.Series)

# Concatenamos las columnas

metdat_florida = pd.concat([metdat_florida, hours_df], axis=1)

metdat_florida = metdat_florida.drop(columns=['hours', 0])

metdat_florida.head(3)
```
