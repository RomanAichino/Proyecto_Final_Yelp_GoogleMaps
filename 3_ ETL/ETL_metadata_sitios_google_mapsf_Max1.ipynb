## 2. Columna Address

#### Filtramos para Florida

```python
print(df_metada.iloc[14]['address'])
print(type(df_metada.iloc[14]['address']))
```

```python
metdat_florida = df_metada[df_metada['address'].str.contains('FL', na=False, case=True)]
metdat_florida = metdat_florida.copy()
metdat_florida.head(3)
```

### Extraemos ciudad(city), stado (states) y codigo postal (postal_code)

```python
regex = r'.+,\s*(?P<city>[^,]+),\s*(?P<states>[^\s]+)\s+(?P<postal_code>\S+)'
new_df = metdat_florida['address'].str.extract(regex) #metdat_florida por fd_metada

metdat_florida = metdat_florida.join(new_df,lsuffix='_original', rsuffix='_new')

metdat_florida.head(3)
```

```python
print(metdat_florida.iloc[11776]['address']) #['address']
print(type(metdat_florida.iloc[11776]['address'])) #['address']
```

```python

# FILTRO POR CODIGO POSTAL

metdat_florida['postal_code'] = pd.to_numeric(metdat_florida['postal_code'], errors='coerce') # metdat_florida

metdat_florida = metdat_florida[(metdat_florida['postal_code'] >= 32003) & (metdat_florida['postal_code'] <= 34997)].copy()
# According to website https://www.zipcode.com.ng/2022/06/list-of-florida-zip-codes-.html
```

```python

# FILTRO POR CIUDADES FLORIDA

df_florida_cities = pd.read_json('/content/drive/MyDrive/Otros/us-cities-table-for-florida.json')
# Fuente Json https://worldpopulationreview.com/states/cities/florida

florida_names_list = df_florida_cities['name'].tolist()

metdat_florida = metdat_florida[metdat_florida['city'].isin(florida_names_list)]
metdat_florida.head(3)
```

## 3. Columna Category

```python
# Le damos el formato adecuado a la columna category:

import numpy as np
import ast

metdat_florida['category'] = metdat_florida['category'].apply(lambda x: ast.literal_eval(x)
                                                                if isinstance(x, str) else np.nan)

def format_category_list(category_list):
    if isinstance(category_list, list):
        return ', '.join(category_list)
    else:
        return category_list

metdat_florida['category'] = metdat_florida['category'].apply(format_category_list)
metdat_florida.head(3)
```

```python
metdat_florida['category'].value_counts().sum()
```

```python
#states_list = ['FL']

#metdat_florida = metdat_florida[metdat_florida['states'].isin(states_list)] # ISIN es un filtro mas especifico(excluye otros terminos y filtra solo para lo indicado)
                                                                              #que CONTAINS(incluye a todos aquellos que tengan la palabra Restaurant por ejemplo)

#business_list = ['Restaurant']

metdat_florida['category'].fillna('', inplace=True)
metdat_florida = metdat_florida[metdat_florida['category'].str.contains('Restaurant')]
#metdat_florida = metdat_florida[metdat_florida['category'].isin(business_list)]
metdat_florida#.head(3)
```

```python
# Realizamos un conteo del top 10 de categorias:

category = metdat_florida['category'].str.split(', ').explode()
category_counts = category.value_counts()/len(metdat_florida)

# Mostrar el conteo de 24 categorías mas importantes
category_counts.head(24)
```

'Planning': ['Quick visit']

```python
metdat_florida['MISC'] = metdat_florida['MISC'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)

desired_keys = ['Service options','Highlights','Popular for','Accessibility','Offerings','Dining options','Crowd','Atmosphere','Payments','Amenities','Service options','Planning']

def extract_key_values(misc_dict, key):
    if isinstance(misc_dict, dict) and key in misc_dict:
        return ', '.join(misc_dict[key])
    else:
        return np.nan

# creamos columnas por cada key
for key in desired_keys:
    metdat_florida[key] = metdat_florida['MISC'].apply(lambda x: extract_key_values(x, key))

# Drop the 'MISC'
metdat_florida = metdat_florida.drop(columns=['MISC'])

metdat_florida
```

```python
columns_to_iterate = ['Service options','Highlights','Popular for','Accessibility','Offerings','Dining options','Crowd','Atmosphere','Payments','Amenities','Service options','Planning']

for col in columns_to_iterate:
    value_counts = metdat_florida[col].value_counts()
    total_count = value_counts.sum()  # Calculate the sum of counts
    print(f"Value Counts for {col}:\n{value_counts}\nTotal Count: {total_count}\n")
```

```python
metdat_florida.drop(columns=['Planning', 'Payments', 'Highlights'], inplace=True)
```

```python
import ast
# De <class 'str'> a <class 'list'>

metdat_florida['hours'] = metdat_florida['hours'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else None)

# De <class 'list'> a <class 'dict'>

metdat_florida['hours'] = metdat_florida['hours'].apply(lambda x: dict(x) if isinstance(x, list) else None)

# De <class 'dict'> a columnas

hours_df = metdat_florida['hours'].apply(pd.Series)

# Concatenamos las columnas

metdat_florida = pd.concat([metdat_florida, hours_df], axis=1)

#metdat_florida = metdat_florida.drop(columns=['hours', 0])

metdat_florida.head(3)
```
